```python
new_df=df[df['subscriber_count']>=1000]
```

```python
new_df=new_df[new_df['mean_views_last_30_videos']>=6000]
```

```python
new_df=new_df.reset_index(drop=True)
```

```python
columns=['channel_id','channel_link','channel_name','description','keywords','avatar','country']
```

```python
new_df=new_df[columns]
```

```python
new_df.dropna(subset=['keywords'],inplace=True)
```

```python
new_df=new_df.reset_index(drop=True)
```

```python
new_df
```

```python
import pandas as pd

def combine_keywords(df, column_name):
    df[column_name] = df[column_name].apply(
        lambda x: ', '.join(keyword.replace(" ", "").lower() for keyword in x.split(','))
    )
    return df
popular_df = combine_keywords(new_df, 'keywords')
```

```python
popular_df=popular_df[popular_df['country']=='India']
popular_df=popular_df.reset_index()
```

```python
import pickle
```

```python
similarity_matrix=pickle.load(open('similarity_matrix.pkl','rb'))
```

```python
vec=pickle.load(open('vectorizer.pkl','rb'))
```

```python
similar_dict=dict()
```

```python
def recommend(channel_name):
#     find the index of the channel name
    indx=np.where(popular_df['channel_name']==channel_name)[0][0]
#     return similar_dict[indx]['similar_channels_indx'],similar_dict[indx]['similarity_score']
    recommended_channels=[]
    scores=similar_dict[indx]['similarity_score']
    for channel in similar_dict[indx]['similar_channels_indx']:
        print(channel)
        data=[]
        name=popular_df.iloc[channel]['channel_name']
        avatar=popular_df.iloc[channel]['avatar']
        link=popular_df.iloc[channel]['channel_link']
        data.append(name)
        data.append(avatar)
        data.append(link)
        recommended_channels.append(data)
    return recommended_channels,scores
```

```python
recommended_channels,_=recommend('Abdul Bari')
```

```python
for i, col in enumerate(recommended_channels):
    print(recommended_channels[i][2])
```

```python
for indx, name in enumerate(popular_df['channel_name']):
    print(indx,name)
    similar_channel_data, similarity_score=recommend(name)
    similar_dict[indx]={"similar_channels_indx":similar_channel_data,"similarity_score":similarity_score}
```
