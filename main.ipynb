```python
import pandas as pd

def combine_keywords(df, column_name):
    df[column_name] = df[column_name].apply(
        lambda x: ', '.join(keyword.replace(" ", " ").lower() for keyword in x.split(','))
    )
    return df
popular_df = combine_keywords(popular_df, 'keywords')
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# Initialize NLTK data
nltk.download('punkt')
nltk.download('stopwords')

def preprocess_text(text):
    """
    Preprocesses the input text by:
    1. Converting to lowercase
    2. Removing punctuation
    3. Removing stopwords
    4. Removing whitespace
    5. Removing links
    6. Applying stemming

    Args:
        text (str): Input text

    Returns:
        str: Preprocessed text
    """

    # Convert to lowercase
    text = text.lower()

    # Remove links
    text = re.sub(r'http\S+', '', text)

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    
    # Remove underscores
    text = re.sub(r'_+', '', text)
    
    # Remove digits and numerical terms
    text = re.sub(r'\d+(k|m|b)?', '', text)

    # Tokenize text
    tokens = word_tokenize(text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    # Apply stemming
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(word) for word in tokens]

    # Join tokens back into a string
    text = ' '.join(tokens)

    return text

# Example usage
text = "   This is a sample text with __links http://example.com and punctuation! Remove stopwords and apply stemming. ðŸ¤” 1.8K"
print(preprocess_text(text))
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

chunk_size = 500
n = X.shape[0]
similarity_matrix = np.zeros((n, n))

for i in range(0, n, chunk_size):
    chunk = X[i:i+chunk_size].toarray()
    similarities = cosine_similarity(chunk, X.toarray())
    similarity_matrix[i:i+chunk_size] = similarities
```

```python
def recommend(channel_name):
#     find the index of the channel name
    indx=np.where(popular_df['channel_name']==channel_name)[0][0]
    similar_channels=sorted(list(enumerate(similarity_matrix[indx])),key=lambda x:x[1],reverse=True)[1:6]
    recommended_channels=[]
    scores=[]
    for channel in similar_channels:
        data=[]
        scores.append(channel[1])
        name=popular_df.iloc[channel[0]]['channel_name']
        avatar=popular_df.iloc[channel[0]]['avatar']
        data.append(name)
        data.append(avatar)
        recommended_channels.append(data)
    return recommended_channels,scores
```
