```python
import pandas as pd
import numpy as np
main_df=pd.read_csv('youtube_channels_1M_clean.csv')
```

```python
new_df=main_df[main_df['subscriber_count']>=1000]
```

```python
new_df=new_df[new_df['mean_views_last_30_videos']>=6000]
```

```python
new_df=new_df.reset_index(drop=True)
```

```python
columns=['channel_id','channel_link','channel_name','description','keywords','avatar','country','total_views']
```

```python
new_df=new_df[columns]
```

```python
new_df.dropna(subset=['keywords'],inplace=True)
```

```python
new_df=new_df.reset_index(drop=True)
```

```python
import pandas as pd

def combine_keywords(df, column_name):
    df[column_name] = df[column_name].apply(
        lambda x: ', '.join(keyword.replace(" ", "").lower() for keyword in x.split(','))
    )
    return df
popular_df = combine_keywords(new_df, 'keywords')
```

```python
popular_df=popular_df[popular_df['country']=='India']
popular_df=popular_df.reset_index()
popular_df
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# Initialize NLTK data
nltk.download('punkt')
nltk.download('stopwords')

def preprocess_text(text):
    """
    Preprocesses the input text by:
    1. Converting to lowercase
    2. Removing punctuation
    3. Removing stopwords
    4. Removing whitespace
    5. Removing links
    6. Applying stemming

    Args:
        text (str): Input text

    Returns:
        str: Preprocessed text
    """

    # Convert to lowercase
    text = text.lower()

    # Remove links
    text = re.sub(r'http\S+', '', text)

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    
    # Remove underscores
    text = re.sub(r'_+', '', text)
    
    # Remove digits and numerical terms
    text = re.sub(r'\d+(k|m|b)?', '', text)

    # Tokenize text
    tokens = word_tokenize(text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    # Apply stemming
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(word) for word in tokens]

    # Join tokens back into a string
    text = ' '.join(tokens)

    return text

# Example usage
text = "   This is a sample text with __links http://example.com and punctuation! Remove stopwords and apply stemming. ðŸ¤” 1.8K"
print(preprocess_text(text))
```

```python
popular_df['keywords_preprocessed']=popular_df['keywords'].apply(preprocess_text)
```

# Let's Cluster channels

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt
# Convert text data into numerical features using TF-IDF
vectorizer = TfidfVectorizer(max_features=12000)
X = vectorizer.fit_transform(popular_df['keywords_preprocessed'])
```

```python
# Determine the optimal number of clusters (Elbow method)
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot the Elbow curve
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
# Silhouette score
n_clusters_list = range(2, 11)
silhouette_scores = []
for i in n_clusters_list:
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    silhouette_avg = silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(silhouette_avg)
plt.plot(n_clusters_list, silhouette_scores)
plt.title('Silhouette Score')
plt.xlabel('Number of clusters')
plt.ylabel('Score')
```

```python
# Choose the optimal number of clusters (e.g., 5)
optimal_clusters = 9

# Perform K-Means clustering
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', random_state=42)
kmeans.fit(X)

# Predict cluster labels
labels = kmeans.labels_

# Print cluster labels
print("Cluster Labels:")
print(labels)
```

```python
random_indices = np.random.choice(X.shape[0], size=int(0.1 * X.shape[0]), replace=False)
# Visualize clusters using PCA (optional)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X[random_indices].toarray())
```

```python
# Visualize clusters
plt.figure(figsize=(8, 6))
for cluster in np.unique(labels[random_indices]):
    cluster_indices = np.where(labels[random_indices] == cluster)[0]
    plt.scatter(X_pca[cluster_indices, 0], X_pca[cluster_indices, 1], label=f'Cluster {cluster}')

plt.title('Cluster Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Clusters', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
